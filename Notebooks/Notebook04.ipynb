### Styles, sizes, legends, labels

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2*np.pi,10)
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
z=np.cos(x)
z
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.plot(x,y,'o')
plt.plot(x,z)
```

```python
plt.plot(x,y,'o',color="red")
```

```python
plt.plot(x,z,color='g',linestyle='--')
```

```python
plt.plot(x,z,'o',color='green')
```

```python
plt.plot(x,z,'r:')
```

```python
plt.plot(x,z,color='green',marker='v',linestyle='')
```

#### Search for *matplotlib* **markers** and **colors**

e.g. [https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.plot.html)

```python
plt.plot(x,z,color='green',marker='v',linestyle='')
plt.plot(x,y,color='k',marker='',linestyle=':')
```

```python
plt.xlim(-4.0, 10.28)    # Set x limits
plt.ylim(-2.0, 2)        # Set y limits
plt.plot(x,y,color='r',marker='',linestyle=':')
```

```python
plt.xticks(np.linspace(0, 2*np.pi, 4, endpoint=True))
plt.plot(x,y,color='r',marker='.',linestyle='',markersize=20)
```

```python
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
          ['$0$', f'$\pi/2$', r'$\pi$', '$3\pi/2$', r'$2\pi$'])
plt.plot(x,y,color='r',marker='',linestyle='--',linewidth=5)
```

```python
plt.figure(figsize=(8,8))
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
          ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$'],fontsize=18)
plt.plot(x,y,color='r',marker='',linestyle='--',linewidth=5)
```

```python
plt.plot(x,y,color='b',label='sin')
plt.plot(x,z,color='r',label='cos')
plt.legend(loc='lower left')
```

```python
plt.plot(x,y,color='b',label='sin')
plt.plot(x,z,color='r',label='cos')
plt.legend(bbox_to_anchor=(1.0, 1.0, 1, 1), loc='center')
```

```python
plt.plot(x,y,color='r',marker='',linestyle='-.',linewidth=5)
plt.xlabel("x")
plt.ylabel('sin(x)',fontsize=20,color='red')
```

```python
plt.plot(x,x**2,color='r',marker='',linestyle='-.',linewidth=5)
plt.xlabel("x")
plt.ylabel('$x^2$',fontsize=20,color='red',rotation=0,labelpad=20)
plt.title('f($x$)= $x^2$',fontsize=30)
```

### Aspect ratio

### Subplots

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x, y, 'r')
plt.subplot(2,2,2)
plt.plot(x, y, 'b')

plt.subplot(2,2,3)
plt.plot(x, z, 'g')
plt.subplot(2,2,4)
plt.plot(x, z, 'y')
```

### Color according to values, scatter plots

```python
xmany=np.linspace(0,2*np.pi,100)
zmany=np.cos(xmany)
plt.scatter(xmany,zmany, c=np.abs(zmany), cmap=plt.cm.get_cmap('hsv'), edgecolor='none')
```

```python
xmany=np.linspace(0,2*np.pi,100)
zmany=np.cos(xmany)
plt.scatter(xmany,zmany, c=(zmany+1)/2.0, cmap=plt.cm.get_cmap('hsv'), edgecolor='none')
```

```python
#plt.scatter(xmany,zmany, c=plt.cm.hot(np.abs(zmany)), edgecolor='none')
plt.scatter(xmany,zmany, c=(zmany+1)/2.0, cmap=plt.cm.get_cmap('hot'), edgecolor='none')
plt.colorbar()
```

## Histograms

```python
n = 10000
mean = 4.0
sigma = .5
bins = 10
gaussian = np.random.normal(mean,sigma,n)
poisson = np.random.poisson(mean,n)
```

```python
gh=plt.hist(gaussian,bins)
```

```python
ph=plt.hist(poisson,bins)
```

```python
gp=[gaussian,poisson]
plt.hist(gp,bins)
```
