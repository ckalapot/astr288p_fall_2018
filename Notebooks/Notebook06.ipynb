## Filling areas between lines.

```python
xx=np.linspace(0,2*np.pi,50)
yy1=xx**2/18
yy2=np.sin(xx)
plt.plot(xx,yy1, color='blue', alpha=1.00)
plt.fill_between(xx,0,yy1, color='blue', alpha=.25)
```

```python
xx=np.linspace(0,2*np.pi,50)
yy1=xx**2/18
yy2=np.sin(xx)
yy3=np.cos(xx)
plt.plot(xx,yy1, color='blue', alpha=1.00)
plt.plot(xx,yy2, color='blue', alpha=1.00)
plt.fill_between(xx,yy1,yy2, color='blue', alpha=.25)
plt.fill_between(xx,yy2,yy3, color='green', alpha=.25)
```
